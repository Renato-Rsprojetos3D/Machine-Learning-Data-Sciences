```python
# Importing Libraries
import pandas as pd
import numpy as np
```

```python
# Loading csv
trans = pd.read_csv('transpose.csv')
trans.head()
```

```python
# Transpose a dataframe
sales = trans.melt(id_vars=['Product','Category'])
sales
```

```python
# Creating a filter in the months of January and February
jan_feb = sales[(sales.variable == 'January')|(sales.variable == 'February')]
jan_feb
```

```python
# Sum up the Sales by Products and Category for January and February months only
jan_feb.groupby(['Product','Category']).sum()
```
